Graph:
A graph is a data structure consisting of two components: nodes (vertices) and edges.

A graph G can be defined as G = (V, E), where V is the set of nodes, and E are the edges between them.

Graph


By definition, a Graph is a collection of nodes (vertices) along with identified pairs of nodes (called edges, links, etc).
In NetworkX, nodes can be any hashable object e.g. a text string, an image, an XML object, another Graph, a customized node object, etc.
(Note: Python’s None object should not be used as a node as it determines whether optional function arguments have been assigned in many functions.)

Multiple Edge Attributes

Create adjacency matrices for multiple edge attributes with structured dtypes:

```python
import numpy as np

G = nx.Graph()
G.add_edge(0, 1, weight=10)
G.add_edge(1, 2, cost=5)
G.add_edge(2, 3, weight=3, cost=-4.0)
dtype = np.dtype([("weight", int), ("cost", float)])
A = nx.to_numpy_array(G, dtype=dtype, weight=None)
print("weight --------")
print(A["weight"])
print("cost --------")
print(A["cost"])
nx.draw(G)
print(G.edges)
print(G.nodes)
```

```python
G = nx.Graph()
G.add_edges_from(
    [
        (0, 1, {'weight': 1.0}),
        (1, 2, {'weight': 2.0}),
        (2, 0, {'weight': 1.0}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[1][2]['weight'])
```

```python
G = nx.MultiGraph()
G.add_edge(0, 1, key='a', weight=1.0)
G.add_edge(0, 1, key='b', weight=5.0)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G)
print(d) 
print(d[0][1]['b']['weight'])
```

```python
G = nx.Graph(
    [
        ("A", "B", {"cost": 1, "weight": 7}),
        ("C", "E", {"cost": 9, "weight": 10}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G)
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("--------------------")
df[["source", "target", "cost", "weight"]]
```

```python
G = nx.MultiGraph([('A', 'B', {'cost': 1}), ('A', 'B', {'cost': 9})])
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['A', 'C'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['B'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
```
